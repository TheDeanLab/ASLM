```python
import image_contrast
from tifffile import imread
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_path = os.path.join("E:", "focus_test", "6x_focus.tif")
image_data = imread(image_path)
PSF_support = 3
z, y, x = np.shape(image_data)
entropy = np.zeros(z)

for i in range(z):
    entropy[i] = image_contrast.normalized_dct_shannon_entropy(image_data[i,:,:], PSF_support, False)
    print("Entropy Value:", entropy[i])
    
```

```python
print(z)
```

```python
x = np.linspace(13400, 13150, 125)
plt.plot(x, entropy)
plt.xlabel('Z-Position')
plt.ylabel('DCT Shannon Entropy')
plt.savefig('6x_focus.pdf', dpi=250, bbox_inches=None, pad_inches=0.4)
plt.show()
```

```python
image_path = os.path.join("E:", "focus_test", "3x_focus.tif")
image_data = imread(image_path)
PSF_support = 3
z, y, x = np.shape(image_data)
entropy = np.zeros(z)

for i in range(z):
    entropy[i] = image_contrast.normalized_dct_shannon_entropy(image_data[i,:,:], PSF_support, False)
    print("Entropy Value:", entropy[i])
    
x = np.linspace(13200, 13600, z)
plt.plot(x, entropy)
plt.xlabel('Z-Position')
plt.ylabel('DCT Shannon Entropy')
plt.savefig('6x_focus.pdf', dpi=250, bbox_inches=None, pad_inches=0.4)
plt.show()
```
